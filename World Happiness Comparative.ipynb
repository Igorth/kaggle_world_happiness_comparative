```python
region_africa = df[df['Region'] == 'Sub-Saharan Africa']
region_africa.head()
```

```python
values_africa = region_africa.groupby('Country')['Happiness Score'].sum()
values_africa.plot(kind='line', rot=45, color='red')

plt.show()
```

```python
sns.swarmplot(x="Region", y="Happiness Score",  data=df)
plt.xticks(rotation=30)

plt.show()
```

```python
w_europe = df[df.Region=='Latin America and Caribbean']
ec_europe = df[df.Region=='Australia and New Zealand']
europe = pd.concat([w_europe,ec_europe],axis=0)
europe.head()
```

```python
sns.lmplot(data=europe,x='Economy (GDP per Capita)',y='Happiness Score',hue="Region")
```
